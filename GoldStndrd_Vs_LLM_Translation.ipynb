```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.translate.bleu_score import corpus_bleu, sentence_bleu, SmoothingFunction
import numpy as np
import warnings

warnings.filterwarnings('ignore')

try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
    nltk.download('punkt_tab')

def evaluate_translation_comprehensive(candidate_text, reference_text):
    """
    Evaluerer oversættelse baseret på 3 metoder:
    1. corpus_bleu (Global brevity penalty)
    2. sentence_bleu (Uden smoothing)
    3. sBLEU (sentence_bleu + method4 smoothing)
    """
    
    # Til corpus_bleu (hele teksten som én enhed)
    cand_full_tokens = candidate_text.split()
    ref_full_tokens = reference_text.split()
    
    # Til sentence_bleu (deler teksten i sætninger)
    cand_sentences = sent_tokenize(candidate_text, language='danish')
    ref_sentences = sent_tokenize(reference_text, language='danish')
    
    min_len = min(len(cand_sentences), len(ref_sentences))
    cand_sent_tokens = [sent.split() for sent in cand_sentences[:min_len]]
    ref_sent_tokens = [sent.split() for sent in ref_sentences[:min_len]]


    #  corpus_bleu()
    # Format: [[reference_tokens]], [candidate_tokens]
    score_corpus = corpus_bleu([[ref_full_tokens]], [cand_full_tokens])
    

    # sentence_bleu() uden smoothing
    scores_sent_no_smooth = []
    for c_tok, r_tok in zip(cand_sent_tokens, ref_sent_tokens):
        # Format: [reference_tokens], candidate_tokens
        score = sentence_bleu([r_tok], c_tok)
        scores_sent_no_smooth.append(score)
        
    avg_sent_no_smooth = np.mean(scores_sent_no_smooth)
    
    # sentence_bleu() med SmoothingFunction().method4 (sBLEU)

    smoother = SmoothingFunction().method4
    scores_sbleu = []
    for c_tok, r_tok in zip(cand_sent_tokens, ref_sent_tokens):
        score = sentence_bleu([r_tok], c_tok, smoothing_function=smoother)
        scores_sbleu.append(score)
        
    avg_sbleu = np.mean(scores_sbleu)

    # Output
    print("--- BLEU Evaluering (NLTK) ---")
    print(f"1. corpus_bleu (Fuld tekst):        {score_corpus:.4f}")
    print(f"2. sentence_bleu (Gns. u. smooth):  {avg_sent_no_smooth:.4f}")
    print(f"3. sBLEU (Gns. m. method4):         {avg_sbleu:.4f}")
    
    print("\nNLTK tilbyder i alt 8 forskellige smoothing-metoder (method1 til method8).")

gold_standard = (
    "Hr. og fru Dursley fra Ligustervænget nummer fire var ganske stolte over, at de var helt og aldeles normale. Det var de sidste mennesker man ville mistænke for at være involveret i mærkelige eller mystiske affærer; de var ganske enkelt af den type mennesker, som ikke tolererer nogen slinger i valsen, hvad ordentlig livsførelse angår. Hr. Dursley var direktør for en stor virksomhed ved navn Grunnings. Den fremstillede drilbor. Han var en stor, tyk mand, næsten uden nogen hals, med et kæmpemæssigt overskæg. Fru Durslev var mager og blond; hendes hals var til gengæld omtrent dobbelt så lang som almindelige menneskers, hvilket var meget praktisk for hende, eftersom hun brugte megen tid på at udspionere sine naboer over havehækken. Parret havde en lille dreng ved navn Dudley. I deres øjne fandtes hans lige ikke i hele den vide verden. Hr. og fru Dursley havde alt hvad de kunne ønske sig, men i tilgift havde de også en hemmelighed, og deres største skræk var at nogen skulle afsløre den."
)

# Model 1's oversættelse
llm_translation = (
    "Herre og fru Dursley, bosat på Privet Drive 4, var stolte over at sige, at de var fuldstændig normale, tak i den grad. De var de sidste mennesker, man ville forvente at blive indblandet i noget underligt eller mystisk, for de stod simpelthen ikke til den slags nonsens. Herr Dursley var direktør for et firma ved navn Grunnings, der fremstillede boremaskiner. Han var en meget stor, kraftig mand med næsten ingen hals, skønt han havde et meget stort overskæg. Fru Dursley var tynd og blond og havde næsten dobbelt så lang hals som almindeligt, hvilket kom hende rigtig til nytte, for hun tilbragte det meste af sin tid med at strække sig over hækken for at spionere på naboerne. Dursley’erne havde en lille søn ved navn Dudley, og efter deres mening var der ingen pænere drenge i hele verden. Dursley’erne havde alt, hvad de ønskede sig, men de havde også en hemmelighed, og deres største frygt var, at nogen ville opdage den."
    )

evaluate_translation_comprehensive(llm_translation, gold_standard)
```
